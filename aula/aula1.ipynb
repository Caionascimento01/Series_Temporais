# Teste de Shapiro

```python
import scipy.stats as stats
```

```python
stats.shapiro(x)
```

```python
e,p=stats.shapiro(x)
p
```

```python
def testenormal(x):
    e,p=stats.shapiro(x)
    if p>=0.05:
        return "A serie é normal"
    else:
        return "A serie NAO é normal"
```

```python
testenormal(x)
```

# teste de estacionaridade

```python
import statsmodels.tsa.stattools 
```

```python
kpss=statsmodels.tsa.stattools.kpss(ser)
```

```python
kpss
```

```python
kpss[3]['5%']
```

```python
kpss[0]
```

```python
def testeestacionaria(x):
    kpss=statsmodels.tsa.stattools.kpss(x)
    teste=kpss[0]
    valorcritico=kpss[3]['5%']
    if teste>valorcritico:
        return "A serie NAO é estacionaria"
    else:
        return "A serie é estacionaria"
```

```python
testeestacionaria(ser)
```

```python
testenormal(ser)
```

# MMS

```python
plt.figure(figsize=(15,5))
plt.plot(series,lw=3,label='Serie')
plt.plot(series.rolling(5).mean(),lw=3,label='Media movel N=5')
plt.plot(series.rolling(12).mean(),lw=3,label='Media movel N=12')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel('QNT',fontsize=20)
plt.grid(alpha=0.2)
plt.legend(fontsize=20)
```

# MME 

```python
series.ewm(alpha=0.5).mean()
```

```python
plt.figure(figsize=(15,5))
plt.plot(series,lw=3,label='Serie')
plt.plot(series.rolling(5).mean(),lw=3,label='Media movel N=5')
plt.plot(series.rolling(12).mean(),lw=3,label='Media movel N=12')
plt.plot(series.ewm(alpha=0.5).mean(),lw=3,label='Media exp = 0.5')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel('QNT',fontsize=20)
plt.grid(alpha=0.2)
plt.legend(fontsize=20)
```
